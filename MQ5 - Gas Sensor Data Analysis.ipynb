### Heatmap 

```python
result = pd.pivot_table(data=df, index='Temp', columns='Humidity',values='R8')
print(result)

#create heat map of temp vs humidity
sns.heatmap(result, annot=True, cmap = 'coolwarm', center=0.117)
plt.show()
```
